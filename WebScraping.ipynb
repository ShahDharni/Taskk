```python
from requests.exceptions import URLRequired

import requests
import csv
from pathlib import Path

def scrape_bid_data():
    page_no = 1
    path_to_file = "stockx.csv"

    unique_urls = set()

    while True:
        print('Hold on creating URL to fetch data...')
        URL = 'https://stockx.com/sneakers?page=' + str(page_no)
        print('URL created: ' + URL)

        scraped_data = requests.get(URL)

        # if scraped_data.status_code != 200:
        #     print(f"Failed to fetch data from {URL}. Exiting...")
        #     break


        if Path(path_to_file).is_file():
            with open(path_to_file, 'r', newline='') as csvfile:
                reader = csv.DictReader(csvfile)
                for row in reader:
                    unique_urls.add(row[URL])

            new_urls = [entry[URL] for entry in scraped_data]

            new_data = [entry for entry in scraped_data if entry[URL] not in unique_urls]

            if new_data:
                with open(path_to_file, 'a', newline='') as csvfile:
                    fieldnames = new_data[0].keys()
                    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
                    for entry in new_data:
                        writer.writerow(entry)
                print("New data appended successfully.")
            else:
                print("No new data to append.")
        else:
            with open(path_to_file, 'w', newline='') as csvfile:
                fieldnames = scraped_data[0].keys()
                writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
                writer.writeheader()
                for entry in scraped_data:
                    writer.writerow(entry)
                print(f"CSV file '{path_to_file}' created with the new data.")

        page_no += 1

scrape_bid_data()
```

```python

from pathlib import Path
import pandas as pd
import requests
from bs4 import BeautifulSoup

base_url = 'https://stockx.com/sneakers?page='
unique_urls = []
path_to_file = "stockx.csv"
path = Path(path_to_file)

for page_no in range(1, 26):
    url = base_url + str(page_no)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    if path.is_file():
        existing_data = pd.read_csv(path_to_file)
        existing_urls = existing_data['URL'].tolist()

        new_urls = [url for url in [base_url + str(page_no) for page_no in range(1, 26)] if url not in existing_urls]

        if new_urls:
            new_data = pd.DataFrame({'URL': new_urls})
            existing_data = existing_data.append(new_data, ignore_index=True)
            existing_data.to_csv(path_to_file, index=False)
            print("New data appended successfully.")
        else:
            print("No new data to append.")
    else:
        new_data = pd.DataFrame({'URL': [url for url in [base_url + str(page_no) for page_no in range(1, 26)]]})
        new_data.to_csv(path_to_file, index=False)
        unique_urls.extend(new_data['URL'].tolist())
        print(f"CSV file '{path_to_file}' created with the new data.")
```
