# **Seaborn**

**What is Seaborn ?**

Seaborn provides a high-level interface to Matplotlib, a powerful but sometimes unwieldy Python visualization library.On Seaborn’s official website, they state:

If matplotlib “tries to make easy things easy and hard things possible”, seaborn tries to make a well-defined set of hard things easy too.

**Features of Seaborn :**

* Using default themes that are aesthetically pleasing.

* Setting custom color palettes.

* Making attractive statistical plots.

* Easily and flexibly displaying distributions.

* Visualizing information from matrices and DataFrames.  

Those last three points are why Seaborn is our tool of choice for Exploratory Analysis. It makes it very easy to “get to know” your data quickly and efficiently.

**Matplotlib Vs Seaborn**


It is summarized that if Matplotlib “tries to make easy things easy and hard things possible”, Seaborn tries to make a well-defined set of hard things easy too.”

Seaborn helps resolve the two major problems faced by Matplotlib; the problems are −

* Default Matplotlib parameters

* Working with data frames

As Seaborn compliments and extends Matplotlib, the learning curve is quite gradual. If you know Matplotlib, you are already half way through Seaborn.

**Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
sns.set(color_codes=True)
%matplotlib inline
```

```python
d=pd.read_csv("Automobile_data.csv")
d.head(10)
```

**Line Charts**

```python
col1=np.linspace(0,10,100)
col2=np.sin(col1)
df=pd.DataFrame({"C1":col1,"C2":col2})
df.head(10)
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8-darkgrid')
%matplotlib inline
sns.lineplot(x=df.C1,y=df.C2)
plt.show()
```

**Adjusting background color using axes.facecolor**

**Changing labels using x.ticks and y.ticks**

```python
sns.axes_style()
```

```python
plt.style.use('seaborn-v0_8-darkgrid')
%matplotlib inline
sns.set(rc={'axes.facecolor':'lightblue', 'axes.grid':True, 'xtick.labelsize':14,'ytick.labelsize':14})
sns.lineplot(x=df.C1,y=df.C2,data=df,linewidth=2.5)
sns.lineplot(x=df.C1,y=df.C2+1,data=df,linewidth=2.5)
plt.show()
```

```python
plt.style.use('seaborn-v0_8-darkgrid')
%matplotlib inline
sns.set(rc={'axes.facecolor':'lightblue', 'axes.grid':True, 'xtick.labelsize':14,'ytick.labelsize':14})
sns.lineplot(x=df.C1,y=df.C2,data=df,linewidth=2.5,color="green")
sns.lineplot(x=df.C1,y=np.cos(df.C2+1),data=df,linewidth=2.5,color="black")
plt.show()
```

**To recover the default matplotlib settings**

```python
mpl.rcParams.update(mpl.rcParamsDefault)
%matplotlib inline
```

```python
plt.figure(figsize=(10,8))
sns.lineplot(x="wheel-base",y="engine-size",data=d)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.lineplot(x="wheel-base",y="engine-size",data=d)
plt.title("Wheel_Base Vs Engine_Size")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.lineplot(x="wheel-base",y="engine-size",data=d,linewidth=5)
plt.title("Wheel_Base Vs Engine_Size")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.lineplot(data=d['wheel-base'],linewidth=3, label="Wheel_Base")
sns.lineplot(data=d['engine-size'],linewidth=3, label="Engine-size")
plt.show()
```

```python
plt.figure(figsize=(10,9))
sns.set(rc={"axes.facecolor":"#283747","axes.grid":False,"xtick.labelsize":10,"ytick.labelsize":10})
plt.title("Engine size",fontsize=10)
plt.xticks(rotation=45)
sns.lineplot(x=d['wheel-base'],y=d['engine-size'].index.values,color="#ff9900")
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.style.use('seaborn-darkgrid')
sns.lineplot(x="wheel-base",y="engine-size",hue="fuel-type",data=d)
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.style.use('seaborn-darkgrid')
sns.lineplot(x="wheel-base", y="engine-size", hue="fuel-type", data=d, markers=True, style="fuel-type")
plt.show()
```

```python
plt.figure(figsize=(14,7))
plt.style.use('seaborn-darkgrid')
sns.lineplot(x="num-of-cylinders", y="engine-size", hue="fuel-type", data=d, markers=True, style="fuel-type",err_style="bars")
plt.show()
```

```python
plt.figure(figsize=(14,7))
plt.style.use('seaborn-darkgrid')
sns.lineplot(x="num-of-cylinders", y="engine-size", hue="fuel-type",data=d)
sns.set(style="dark",)
plt.show()
```

```python
plt.figure(figsize=(14,7))
sns.set(rc={'xtick.labelsize':14, "ytick.labelsize":10, "axes.labelsize":15})
sns.lineplot(x="wheel-base", y="engine-size", style="fuel-type", hue="fuel-type", dashes=False, palette = 'CMRmap', markers=["o", ">"] ,data=d)
plt.show()
```

```python
plt.figure(figsize=(16,16))
plt.style.use("seaborn-darkgrid")
plt.gcf().text(.2,.70,"Line Plot",fontsize=35,color="blue",ha="center",va="center")
sns.lineplot(x="wheel-base",y="engine-size",hue="num-of-cylinders",style="fuel-type",data=d)
plt.show()
```
