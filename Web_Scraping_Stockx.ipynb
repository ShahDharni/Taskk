```python
import requests
```

```python
import pandas as pd
```

```python
from bs4 import BeautifulSoup
```

```python
url="https://stockx.com/sneakers?page=1"
```

```python
response=requests.get(url)
```

```python
response = response.content
```

```python
soup=BeautifulSoup(response,'html.parser')
```

```python
soup
```

```python
results=soup.find_all('div',class_="css-1yh5062")
```

```python
shoes=[]

for result in results:

  image=result.find('img')
  title=image.attrs['alt']
  price=result.find('p',class_="chakra-text css-nsvdd9").text
  # price=price[1:]
  ask=result.find('p',class_="chakra-text css-jsq4f").text
  shoes.append([title,price,ask])
  # print(image)
  # print(title)
  # print(price)
  # print(ask)
print(shoes)
  # print()
```

```python
df=pd.DataFrame(shoes,columns=['Title','Price','Ask'])
```

```python
df.to_csv('stockx.csv')
```
