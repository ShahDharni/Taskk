## **To Scrap Internal infomation**
So we are going to open internal links of the website inorder to extract more information

````python
import re

baseUrl = 'http://books.toscrape.com/catalogue/'
allPages = ['http://books.toscrape.com/catalogue/page-1.html',
            'http://books.toscrape.com/catalogue/page-2.html']

for url in allPages:
    res = requests.get(url)
    soup = BeautifulSoup(res.text,'html.parser')
    article = soup.find_all(class_ = 'product_pod')
    f = len(article)
    perBook=[]

    for i in range(f):
        perBook.append(baseUrl + article[i].h3.a.get('href'))
        column_names = ['Title', 'Link', 'Price', 'Quantity in Stock']
        data  =  []

    i = 0
    for url in perBook:
        res = requests.get(url)
        soup = BeautifulSoup(res.text,'html.parser')
        table = soup.find_all('tr')
        #for i in table:
        #    print(i.text.strip())
        title = soup.h1.text
        price = table[2].text.strip()[-5:]
        x = table[5].text.strip()
        quantity = re.sub(r'[!A-Za-z()]','', x).strip()
        data.append([title, url, price, quantity])
        print(title , " " , url , " " , price , " " , quantity)
        i+=1
````
